```python
import numpy as np
```

```python
arr1 = np.array([[1, 2, 3, 4]])
arr2 = np.array([[2, 2, 2, 2]])

arr3 = np.concatenate((arr1, arr2))
print(arr3)
```

```python
arr3 = np.concatenate((arr1, arr2), axis = 1)
arr3
```

```python
arr4 = np.array([[4, 5, 6, 7], [9, 9, 9, 9]])
arr5 = np.array([[4, 5, 6, 7], [1, 1, 1, 1]])
arr6 = np.concatenate((arr4, arr5), axis=0)
arr6
```

```python
arr7 = np.concatenate((arr4, arr5), axis=1)
arr7
```

```python
arr8 = np.vstack((arr4, arr5))
arr8
```

```python
arr9 = np.hstack((arr1, arr2))
arr9
```

```python
arr10 = np.array([[1, 2, 3], [4, 5, 6]])
arr11 = np.array([[1, 2, 3]])
# np.hstack((arr10, arr11)) ### NOT Possible
np.vstack((arr10, arr11))
```
