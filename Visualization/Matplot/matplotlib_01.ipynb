### Histogram

```python
x = np.random.normal(50, 2.5, 100)
x
```

```python
plt.hist(x, bins=10)
plt.show()
```

```python
language = ['c++', 'python', 'c#', 'java', 'rust']
popularity = [56, 89, 67, 60, 34]
explode = [0.2, 0.5, 0.4, 0.3, 0.1]
plt.pie(popularity, labels=language, autopct="%.2f%%", explode=explode)
plt.show()
```
