```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv(r'iris.csv')
```

```python
iris.head()
```

## getting the basic info about the **data**

```python
iris.info()
```

## Distribution plot

```python
# Histogram

sns.histplot(data = iris, x = 'petal_width', bins=20, hue = "species")
plt.title("Distribution of petal width")
plt.show()
```

```python
sns.histplot(data = iris, x = 'sepal_width', bins=20, hue = "species")
plt.title("Distribution of sepal width")
plt.show()
```

```python
sns.histplot(data = iris, x = 'sepal_length', bins=20, hue = "species")
plt.title("Distribution of sepal length")
plt.show()
```

## Barplot

```python
sns.barplot(data = iris, x = "species", y = 'sepal_length', palette = 'Set2')
plt.title("Bar Graph Distribution of sepal_length based on species")
plt.show()
```

```python
sns.barplot(data = iris, x = "species", y = 'sepal_width', palette = 'Set2')
plt.title("Bar Graph Distribution of sepal_width based on species")
plt.show()
```

# Scatter plot

```python
iris.head()
```

```python
sns.scatterplot(data = iris, x = "sepal_length", y = 'petal_length', hue = 'species')
plt.title("scatter Distribution of sepal_width based on species")
plt.show()
```

```python
sns.scatterplot(data = iris, x = "sepal_width", y = 'petal_length', hue = 'species')
plt.title("scatter Distribution of sepal_width based on species")
plt.show()
```

## Box Plot

```python
sns.boxplot(data = iris, x = "species", y = 'sepal_length')
plt.title("Boxplot shoeing the Distribution of sepal_length")
plt.show()
```
